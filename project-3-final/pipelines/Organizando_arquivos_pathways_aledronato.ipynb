```python
import requests
import pandas as pd
import os
from time import sleep

# Fun√ß√£o para buscar o Entrez ID a partir do UniProt ID
def buscar_entrez_por_uniprot(uniprot_id):
    url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}.json"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        # Buscar Entrez Gene ID (GeneID) entre os cross-references
        for xref in data.get("uniProtKBCrossReferences", []):
            if xref["database"] == "GeneID":
                return xref["id"]
        return None
    except Exception:
        return None

# Fun√ß√£o para processar um CSV inteiro
def adicionar_entrez_ao_csv(caminho_csv, coluna_uniprot="Identifier"):
    print(f"\nüîπ Processando arquivo: {caminho_csv}")

    # L√™ o CSV
    df = pd.read_csv(caminho_csv)

    # Cria lista de UniProt IDs √∫nicos
    uniprot_ids = df[coluna_uniprot].dropna().unique().tolist()

    # Dicion√°rio de mapeamento UniProt -> Entrez
    mapeamento = {}
    for uniprot_id in uniprot_ids:
        entrez_id = buscar_entrez_por_uniprot(uniprot_id)
        mapeamento[uniprot_id] = entrez_id
        print(f"{uniprot_id} ‚Üí {entrez_id}")
        sleep(0.5)  # evita sobrecarregar o servidor da UniProt

    # Adiciona a nova coluna ao DataFrame
    df["Entrez_ID"] = df[coluna_uniprot].map(mapeamento)

    # Gera nome de sa√≠da
    nome_base, ext = os.path.splitext(caminho_csv)
    novo_arquivo = f"{nome_base}_com_entrez{ext}"

    # Salva o novo CSV
    df.to_csv(novo_arquivo, index=False)
    print(f"‚úÖ Arquivo salvo como: {novo_arquivo}\n")

# ---- Lista dos seus arquivos ----
arquivos = [
    "cytokine_signaling.csv",
    "ESR_signaling.csv",
    "G_alpha_q_signaling_events.csv",
    "Interferon_alpha_beta.csv",
    "Interferon_gamma.csv"
]

# ---- Executa para todos ----
for arquivo in arquivos:
    adicionar_entrez_ao_csv(arquivo)
```
