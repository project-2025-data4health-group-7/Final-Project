#### Genes em comum entre a expressão diferencial e os targets do medicamento

```python
import pandas as pd

via_hippo= pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Hippo signaling regulation.tsv', sep='\t')
via_hippo['gene'] = via_hippo['Label']

via_ESR = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\ESR-mediated signaling.tsv', sep='\t')
via_ESR[['uniprot', 'gene']] = via_ESR['MoleculeName'].str.split(' ', n=1, expand=True)

via_nuclear = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Signaling by Nuclear Receptors.tsv', sep='\t')
via_nuclear[['uniprot', 'gene']] = via_nuclear['MoleculeName'].str.split(' ', n=1, expand=True)

via_insulina = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Signaling by Insulin receptor.tsv', sep='\t')
via_insulina = via_insulina[via_insulina['MoleculeType'] == 'Proteins']
via_insulina[['uniprot', 'gene']] = via_insulina['MoleculeName'].str.split(' ', n=1, expand=True)

via_tyrosine = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Signaling by Receptor Tyrosine Kinases.tsv', sep='\t')
via_tyrosine[['uniprot', 'gene']] = via_tyrosine['MoleculeName'].str.split(' ', n=1, expand=True)

via_cardio = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Diabetic cardiomyopathy.csv', header=None).T
via_cardio.columns = ['gene'] 

via_sphingo = pd.read_csv('project-2\\Genes-Alvo-dos-Medicamentos\\acido_zoledronico\\vias\\Sphingolipid signaling pathway.csv', header=None).T
via_sphingo.columns = ['gene'] 
```

```python
for via in [via_sphingo, via_cardio, via_ESR, via_hippo, via_insulina, via_nuclear, via_tyrosine]:
    genes_vias = set(via['gene'].tolist())
    comuns = set(genes_vias).intersection(alvos_ac_zoled['gene'].tolist())
    print(comuns)
```

```python

for GSE in [7158, 56814, 56815]:
    print(F'Dataset GSE{GSE}')
    genes = pd.read_csv(f'project-2\Resultados-de-expressao-diferencial\GSE{GSE}\gene_nodes_cytoscape.csv')
    display(genes)
    alvos_gerais = set()
    for via in [via_sphingo, via_cardio, via_ESR, via_hippo, via_insulina, via_nuclear, via_tyrosine]:


        genes_vias = set(via['gene'].tolist())

        comuns = set(genes['preferredName']).intersection(genes_vias)

        print(f'\tGenes em comum com a via: {comuns}')

        lista_comuns = set(comuns)

        alvos_gerais.update(lista_comuns)

    print(f'\n\tMatches entre vias e conjunto de dados GSE{GSE}: {alvos_gerais}\n')

    genes['alvo_ac_zol'] = pd.Series([0]*len(genes_vias))
    genes.loc[genes['preferredName'].isin(alvos_gerais), 'alvo_ac_zol'] = 1
    genes.to_csv(f'project-2\Resultados-de-expressao-diferencial\GSE{GSE}\gene_nodes_cytoscape_alvos_GSE{GSE}.csv', mode='w', header=True, index=False)
```
