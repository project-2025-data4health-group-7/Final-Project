### Filtragem dos Dados

```python
# Quando a probabilidade pra um modelo é igual a 1, para todos os outros também é
dados_filtrados = data[data['GM'] == 1]
```

### Busca pela proteína correspondente ao ID do Uniprot

```python
import requests
import sys
import os

def encontrar_gene_por_id(uniprot_id, output_path = 'lista_genes.csv'):
   
    url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}"

    params = {
        "fields": "gene_names,protein_name",
        "format": "json"
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  

        data = response.json()

        if "genes" in data and len(data["genes"]) > 0:
            nome_gene = data["genes"][0]["geneName"]["value"]
            nome_proteina = data["proteinDescription"]["recommendedName"]["fullName"]["value"]

            dados_extraidos = {
                'proteína': nome_proteina,
                'gene': nome_gene
            }

            df_dados = pd.DataFrame([dados_extraidos])
            escrever_cabecalho = not os.path.exists(output_path)
            df_dados.to_csv(output_path, mode='a', header=escrever_cabecalho, index=False)

            return f"O UniProt ID '{uniprot_id}' corresponde à proteína '{nome_proteina}' e o nome do gene é: {nome_gene}"
        else:
            return f"UniProt ID '{uniprot_id}' encontrado, mas não há um nome de gene associado no registro."

    except requests.exceptions.HTTPError as e:
        if e.response.status_code == 404:
            return f"Nenhum registro encontrado para o UniProt ID: {uniprot_id}"
        else:
            return f"Ocorreu um erro HTTP ao buscar os dados: {e}"
    except requests.exceptions.RequestException as e:
        return f"Ocorreu um erro ao conectar com o UniProt: {e}"
    except Exception as e:
        return f"Ocorreu um erro inesperado: {e}"

```

```python
i = 0
for uid in dados_filtrados['uID']:
    i+=1
    print(f'Buscando ID {uid} no UniProt ({i}/{len(dados_filtrados["uID"])})')
    encontrar_gene_por_id(uid)
```
